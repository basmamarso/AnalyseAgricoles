```python
# EDA for Historique_rendements

## Loading Data

import pandas as pd

# Load data
df = pd.read_csv("../data/historique_rendements.csv")

# Display first rows
df.head()
```

```python
# Check columns and data types
df.info()

# Check for missing values
df.isnull().sum()

# Basic statistics
df.describe()

## Visualizations

import matplotlib.pyplot as plt
import seaborn as sns

# Plot distributions
df.hist(figsize=(10, 8), bins=30)
plt.show()
# Slectionner uniquement les colonnes numriques
numeric_df = df.select_dtypes(include=['float64', 'int64'])
if numeric_df.shape[1] > 0:  # Vrifier si des colonnes numriques existent
    plt.figure(figsize=(10, 6))
    sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
    plt.show()
else:
    print("Aucune colonne numrique disponible pour la heatmap.")
```
